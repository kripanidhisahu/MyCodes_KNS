```python
#Import relevant libraries
from pandas import read_csv
import pandas as pd
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster 
# import warnings
# warnings.filterwarnings("ignore")
```

```python
wine = pd.read_csv('D:/1.0 Studies/DataScience_Excelr/Assignment12_PCAtSNE/wine.csv')
wine.head()
```

```python
# Split-out validation dataset
array = wine.values
# separate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
#TSNE visualization
from bioinfokit.visuz import cluster

data_tsne = TSNE(n_components=2).fit_transform(X)
cluster.tsneplot(score=data_tsne)
```

```python
# get a list of categories
color_class = wine['Type'].to_numpy()
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15, 1))

#Plot will be stored in the default directory
```
