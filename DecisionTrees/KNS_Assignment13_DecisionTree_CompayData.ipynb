### Decision Tree - Assignment


About the data: 
A Company dataset with around 10 variables and 400 records. 

The attributes are as follows: 

Sales -- Unit sales (in thousands) at each location

Competitor Price -- Price charged by competitor at each location

Income -- Community income level (in thousands of dollars)

Advertising -- Local advertising budget for company at each location (in thousands of dollars)

Population -- Population size in region (in thousands)

Price -- Price company charges for car seats at each site

Shelf Location at stores -- A factor with levels Bad, Good and Medium indicating the quality of the shelving location for the 
car seats at each site

Age -- Average age of the local population

Education -- Education level at each location

Urban -- A factor with levels No and Yes to indicate whether the store is in an urban or rural location

US -- A factor with levels No and Yes to indicate whether the store is in the US or not
  
### Problem Statement

A cloth manufacturing company is interested to know about the segment or attributes that cause high sale. 

Approach: A decision tree can be built with target variable as 'Sales' (we shall convert it into a categorical variable). All other variable remain independent in the analysis.

```python
company_data1=pd.get_dummies(company_data,columns=['ShelveLoc','US','Urban'],drop_first = True)
company_data1
```

```python
#Segregating the sales to low / high between ranges [2,9,15] and assiging Yes/No
company_data1["Target_Sales"] = pd.cut(company_data1["Sales"], bins = [2,9,15], labels = ["No", "Yes"])
company_data1
```

```python
#Creating dummies for the target sales
company_data2 = pd.get_dummies(company_data1,columns = ["Target_Sales"],drop_first=True)
company_data2.head(10)
```

```python
#checking for multicollinearity problem using VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["Variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = company_data2.iloc[:,1:]
calc_vif(X)
```

CompPrice, Price, Education have high multicollenierity. These variables can be dropped from the analysis.

```python
#Correlation check
company_data2.corr()
```

```python
company_data3=company_data2.drop(['CompPrice','Price','Education','Age','Income'], axis=1)
company_data3
```

```python
# Converting Sales 
company_data3['Sales'] = company_data3['Sales'].astype('category')
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
company_data3_norm = norm_func(company_data3.iloc[:,1:])
company_data3_norm.head(10)
```

```python
# Declaring features & target (Another Method)
#X = company_data3_norm.drop(['Target_Sales_Yes'], axis=1)
#y = company_data3_norm['Target_Sales_Yes']
```

```python
# Declaring features & target
X=company_data3_norm.iloc[:,0:6]
y=company_data3_norm['Target_Sales_Yes']
```

```python
# Splitting data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
x_train.isnull().sum()
```

### Building Decision Tree Classifier using Entropy Criteria

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
#Plot the decision tree
tree.plot_tree(model);
```

```python
fn=['Advertising','Population','ShelveLoc_Good','ShelveLoc_Medium','US_Yes','Urban_YES']
cn=['0', '1']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (6,4), dpi=300)
tree.plot_tree(model,
               feature_names = fn,
               class_names=cn,
               filled = True);
```

```python
#Prediction using test data
preds = model.predict(x_test) # predicting on test data set 
pd.Series(preds).value_counts() # getting the count of each category
```
